```python
import requests
from bs4 import BeautifulSoup
page = requests.get("https://stackoverflow.com/questions")
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
p=soup.select(".question-summary")
#question its answer and the no of votes and views are extacted one by one from soup
for i in p:
    print(i.select_one('.question-hyperlink').getText())
    print(i.select_one('.excerpt').getText())
    print(i.select_one('.vote-count-post').getText())
    print(i.select_one('.views').getText())
```
